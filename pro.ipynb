```python
import requests
import csv
```

```python
r = requests.get('https://reqres.in/api/users')
r
```

```python
data = r.json()["data"]
data
```

```python
type(data)
```

```python
with open('reqres.csv', 'w', newline= '') as file:

    # Create Columns names
    writer = csv.DictWriter(file, fieldnames= ['id', 'first_name', 'last_name', 'email', 'avatar'])
    writer.writeheader()
    for user in data:
        id = user['id']
        first_name = user['first_name']
        last_name =  user['last_name']
        email =  user['email']
        avatar =  user['avatar']

    writer.writerow({'id' : id, 'first_name' : first_name, 'last_name' : last_name, 'email' : email, 'avatar' : avatar})    
        
```
